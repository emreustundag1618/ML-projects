```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/Billionaire.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace = True)
```

```python
data.info()
```

```python
# NetWorth column object to float
data["NetWorth"] = data["NetWorth"].str.strip("$")
data["NetWorth"] = data["NetWorth"].str.strip("B")
data["NetWorth"] = data["NetWorth"].astype(float)
```

```python
data.head()
```

```python
# the first 10 richest people
df_10 = data.sort_values(by = ["NetWorth"], ascending = False).head(10)
```

```python
plt.figure(figsize = (20, 10))
sns.histplot(x = "Name", hue = "NetWorth", data = df_10)
plt.show()
```

```python
# first most number of source
a = data["Source"].value_counts().head()
```

```python
# display a pie chart according to sources
index = a.index
values = a.values
custom_colors = ["skyblue","yellowgreen","tomato","blue","red"]
plt.figure(figsize = (5,5))
plt.pie(values, labels = index, colors = custom_colors)
central_circle = plt.Circle((0,0), 0.5, color = "white")
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc("font", size = 12)
plt.title("Top 5 Domains to Become a Billionaire", fontsize = 20)
plt.show()
```
